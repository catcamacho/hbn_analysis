```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(context='talk', style='white')

data = pd.read_csv('movieDM_regressors_setA.csv', index_col=0)
data.index.name = 'TR'
data['seconds'] = data.index * 0.8
data.index = data['seconds']
```

```python
data[['pos_imca','neg_imca']].plot(subplots=True, figsize=(12,5), xlim=(0,600), color=['darkblue','darkred'], 
                                   legend=False, xlabel='Time (seconds)', ylabel='Intensity')
plt.tight_layout()
plt.savefig('movieDM_posnegtraces.png')
```
